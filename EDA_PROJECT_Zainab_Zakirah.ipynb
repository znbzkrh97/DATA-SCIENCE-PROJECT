```python
# Step 10: Data Transformation
#Price and Kilometer-Driven variables are highly skewed and on a larger scale. Lets do log transformation.
#Log transformation can help in normalization, so this variable can maintain standard scale with other variables:

# Function for log transformation of the column
def log_transform(used_cars,col):
  for colname in col:
    if (used_cars[colname] == 1.0).all():
      used_cars[colname + '_log'] = np.log(used_cars[colname]+1)
    else:
      used_cars[colname + '_log'] = np.log(used_cars[colname])
  used_cars.info()
```

```python
# Step 14: Impute Missing Values

#- Imputation Methods Vary: Missing data can be handled using simple techniques like mean, median, mode, or zero imputation, as well as advanced methods like KNN or regularization.
#- Domain Knowledge Matters: Effective imputation requires understanding the business context or data characteristicsblindly applying statistical methods may lead to misleading results.
#- Mileage Imputation Strategy: In the dataset, missing or zero Mileage values (likely due to data entry errors) were replaced with zero initially, then imputed using the mean since it closely matches the median.

# We observed earlier some observations have zero Mileage. This looks like a data entry issue.
# We can fix this by filling null values with zero and then using the mean value of Mileage, since the mean and median values are nearly the same for this variable.
# Therefore, you choose the mean to impute the values.

used_cars.loc[used_cars["Mileage"]==0.0,'Mileage']=np.nan
used_cars.Mileage.isnull().sum()
```

```python
used_cars['Mileage'] = used_cars['Mileage'].astype(str).str.replace(' km/kg', '', regex=False)
used_cars['Mileage'] = used_cars['Mileage'].astype(str).str.replace(' kmpl', '', regex=False)
used_cars['Mileage'] = pd.to_numeric(used_cars['Mileage'], errors='coerce')

used_cars['Mileage'].fillna(value=np.mean(used_cars['Mileage']),inplace=True)
```

```python
used_cars.Seats.isnull().sum()
used_cars['Seats'].fillna(value=np.nan,inplace=True)
used_cars['Seats']=used_cars.groupby(['Model','Brand'])['Seats'].transform(lambda x:x.fillna(x.median()))

used_cars['Engine'] = used_cars['Engine'].astype(str).str.replace(' CC', '', regex=False)
used_cars['Power'] = used_cars['Power'].astype(str).str.replace(' bhp', '', regex=False)
used_cars['Engine'] = pd.to_numeric(used_cars['Engine'], errors='coerce')
used_cars['Power'] = pd.to_numeric(used_cars['Power'], errors='coerce')

used_cars['Engine']=used_cars.groupby(['Brand','Model'])['Engine'].transform(lambda x:x.fillna(x.median()))
used_cars['Power']=used_cars.groupby(['Brand','Model'])['Power'].transform(lambda x:x.fillna(x.median()))

print("Missing values after imputation:")
print(used_cars[['Seats', 'Engine', 'Power']].isnull().sum())
```
